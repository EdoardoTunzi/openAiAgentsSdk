```python
from pydantic import BaseModel, Field

class MessageOutput(BaseModel): 
    response: str = Field(description="Response to the user question")
    explanation: str = Field(description="Explanation for the given answer")


agent = Agent(name="Assistant", instructions="You are a helpful assistant", model="gpt-4.1-nano", output_type=MessageOutput)

result = await Runner.run(agent, "Write a haiku about recursion in programming.")

print(result.final_output)
```

```python
print(result.final_output.response)
print("---")
print(result.final_output.explanation)
```

```python
class User(BaseModel):
    id: int = Field(description="user identification number")
    name: str = Field(description="user full name and surname if both are present")
    mail: str = Field(description="user mail address")

agent = Agent(name="Assistant", instructions="Analize the given text and extract the main information", model="gpt-4.1-nano", output_type=User)

query = "Oggi Mario Rossi con il badge numero 12345 ha  mandato una email da mario.rossi@gmail.com"

result = await Runner.run(agent, query)

print(f"ID: {result.final_output.id}")
print(f"Name: {result.final_output.name}")
print(f"Mail: {result.final_output.mail}")
```
