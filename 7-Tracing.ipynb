### Basic Agent Run

```python
from agents import Agent, Runner
import asyncio

triage_agent = Agent(
    name = "Basic Agent",
    instructions= "You are a useful assistant",
    model= "gpt-4.1-nano"
)

result = await Runner.run(triage_agent, "Ciao, io sono Simone")
print(result.final_output)
```

### Simple function tool calling

```python
from agents import function_tool


@function_tool  
async def get_user_by_id(user_id:str) -> str:
    """Restituisce il nome dell'utente a partire dal suo ID"""
    if user_id == "0000":
        return "Mario Rossi"

    elif user_id == "0001":
        return "Lucia Bianchi"

    elif user_id == "0002":
        return "Carlo Verdi"

    else:
        return "Utente sconosciuto"


triage_agent = Agent(
    name="Agent with tools",
    instructions="You are a useful assistant. You can use the available tools to manage user questions",
    tools=[get_user_by_id],
    model="gpt-4.1-nano"
)

result = await Runner.run(triage_agent, "In uffucio ho trovato un badge con id 0001, a chi appartiene?")
print(result.final_output)
```

### Handoff Diretto

```python
history_tutor_agent = Agent(
    name="History Tutor",
    handoff_description="Specialist agent for historical questions",
    instructions="You provide assistance with historical queries. Explain important events and context clearly.",
    model="gpt-4.1-nano"
)

philosophy_tutor_agent = Agent(
    name="Philosophy Tutor",
    handoff_description="Specialist agent for philosophy questions",
    instructions="You provide help with philosophy questions. Discuss key thinkers, arguments, and schools of thought with clarity. Encourage reflection and critical thinking.",
    model="gpt-4.1-nano"
)

triage_agent = Agent(
    name="Triage Agent - Handoff Diretto",
    instructions="Determine which agent to use based on the user's question",
    handoffs=[history_tutor_agent, philosophy_tutor_agent],
    model="gpt-4.1-nano"
)


result = await Runner.run(triage_agent, "Cosa è la filosofia?")
print(result.final_output)
print("---")
print(result.last_agent.name)
```

### Agents as Tools

```python
history_tool = history_tutor_agent.as_tool(
    tool_name="ask_historical_agent",
    tool_description="Return an answer to historical questions"
)

philosophy_tool = philosophy_tutor_agent.as_tool(
    tool_name="ask_philosophical_agent",
    tool_description="Return an answer to philosophical questions"
)

triage_agent = Agent(
    name="Triage Agent - Agents as Tools",
    instructions="You are a useful assistant. You must use the available tools to manage user questions",
    tools=[history_tool, philosophy_tool],
    model="gpt-4.1-nano"
)

result = await Runner.run(triage_agent, "Cosa è la filosofia?")
print(result.final_output)
print("---")
print(result.last_agent.name)
```
