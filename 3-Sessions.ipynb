### Senza memoria conversazionale

```python
from agents import Agent, Runner
```

```python
agent = Agent(
    name="Assistant",
    instructions="Reply very concisely.",
)
```

```python
result = await Runner.run(
    agent,
    "Ciao, io sono Simone",
)
print(result.final_output)  
```

```python
result = await Runner.run(
    agent,
    "Sai dirmi come mi chiamo?",
)
print(result.final_output)  
```

### Aggiunta di memoria conversazionale

```python
from agents import SQLiteSession
```

```python
# Create a session instance with a session ID
session = SQLiteSession("conversation")
```

```python
result = await Runner.run(
    agent,
    "Ciao, io sono Simone",
    session=session
)
print(result.final_output)  
```

```python
result = await Runner.run(
    agent,
    "Sai dirmi come mi chiamo?",
    session=session
)
print(result.final_output)  
```

### Multiple sessions

```python
session_a = SQLiteSession("conversation_A")

session_b = SQLiteSession("conversation_B")
```

```python
result = await Runner.run(
    agent,
    "Ciao, io sono Simone",
    session=session_a
)
print(result.final_output)  
```

```python
result = await Runner.run(
    agent,
    "Ciao, io sono Enzo",
    session=session_b
)
print(result.final_output)  
```

```python
result = await Runner.run(
    agent,
    "Sai dirmi come mi chiamo?",
    session=session_a
)
print(result.final_output)
```

```python
result = await Runner.run(
    agent,
    "Sai dirmi come mi chiamo?",
    session=session_b
)
print(result.final_output)
```
