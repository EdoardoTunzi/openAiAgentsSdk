```python
from fastmcp import Client
import asyncio
from agents import Agent, Runner
from agents.mcp import MCPServerStdio
```

```python
mcp_servers = {
    "myServerStdio": {         
        "args": ["server_stdio.py"],
        "command": "python",
        "transport": "stdio",
    },
}
```

```python
async with Client(mcp_servers) as client:
    tools = await client.list_tools()
    for tool in tools:
        print(f"{tool.name}: {tool.description} \n")
```

```python
async with Client(mcp_servers) as client:    
    res = await client.call_tool("get_user_by_id", {"user_id":"0001"})
    print(res.content[0].text)
```

```python
async with Client(mcp_servers) as client:    
    res = await client.call_tool("is_prime", {"n":21})
    print(res.content[0].text)
```

```python
async with Client(mcp_servers) as client:    
    res = await client.call_tool("prompt_creative_writing", {"topic":"drago"})
    print(res.content[0].text)
```

```python
async def run_personal_assistant(prompt):
    
    async with MCPServerStdio(
        params={"command": "python", "args": ["server_stdio.py"]},
        name="myServerStdio",
        client_session_timeout_seconds=30,
    ) as mcp_server:

        agent = Agent(
            name="AssistentePersonale",
            instructions="Sei un assistente personale che aiuta l'utente. Usa i tool disponibili se necessario.",
            mcp_servers=[mcp_server],
            model="gpt-4o",
        )

        result = await Runner.run(starting_agent=agent, input=prompt)
        return result.final_output

output = await run_personal_assistant("Il numero 21 è un numero primo?")
print(output)
```
