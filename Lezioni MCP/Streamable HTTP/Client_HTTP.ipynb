```python
from fastmcp import Client
import asyncio
from agents import Agent, Runner
from agents.mcp import MCPServerStreamableHttp
```

```python
DEFAULT_URL = "http://127.0.0.1:8000/mcp"
```

```python
async with Client(DEFAULT_URL) as client:
    tools = await client.list_tools()
    for tool in tools:
        print(f"{tool.name}: {tool.description} \n")
```

```python
async with Client(DEFAULT_URL) as client:    
    res = await client.call_tool("get_user_by_id", {"user_id":"0001"})
    print(res.content[0].text)
```

```python
async with Client(DEFAULT_URL) as client:    
    res = await client.call_tool("is_prime", {"n":17})
    print(res.content[0].text)
```

```python
async with Client(DEFAULT_URL) as client:    
    res = await client.call_tool("prompt_creative_writing", {"topic":"samurai"})
    print(res.content[0].text)
```

```python
async def run_personal_assistant(prompt):
    
    async with MCPServerStreamableHttp(
            params={
                "url": DEFAULT_URL,
                "timeout": 30.0
            },
            cache_tools_list=False,
            name="myServerHTTP",
            client_session_timeout_seconds=30
        ) as mcp_server:

        agent = Agent(
            name="AssistentePersonale",
            instructions=(
                "Sei un assistente personale che aiuta l'utente"
                "Usa i tool disponibili se necessario"),
            mcp_servers=[mcp_server],
            model="gpt-4o"
        )

        result = await Runner.run(starting_agent=agent, input=prompt)
        return result.final_output


output = await run_personal_assistant("In ufficio ho trovato il badge con ID 0002, di chi è?")
print(output)
```
