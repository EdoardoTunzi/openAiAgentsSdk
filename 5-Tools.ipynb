```python
from agents import Agent, Runner, function_tool
import asyncio
```

```python
@function_tool  
async def get_user_by_id(user_id:str) -> str:
    """Restituisce il nome dell'utente a partire dal suo ID"""
    if user_id == "0000":
        return "Mario Rossi"

    elif user_id == "0001":
        return "Lucia Bianchi"

    elif user_id == "0002":
        return "Carlo Verdi"

    else:
        return "Utente sconosciuto"
```

```python
triage_agent = Agent(
    name="Triage Agent",
    instructions="You are a huseful assistant. You can use the available tool to manage user questions",
    tools=[get_user_by_id],
    model="gpt-4.1-nano"
)
```

```python
result = await Runner.run(triage_agent, "Ciao, come puoi aiutarmi?")
print(result.final_output)
print("---")
print(result.last_agent.name)
```

```python
result = await Runner.run(triage_agent, "In uffucio ho trovato un badge con id 0001, a chi appartiene?")
print(result.final_output)
print("---")
print(result.last_agent.name)
```
