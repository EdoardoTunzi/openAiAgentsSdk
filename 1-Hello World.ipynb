### OpenAI Key

##### Creazione della chiave
Per utilizzare i modelli di OpenAI dobbiamo avere una chiave API valida che possiamo creare su https://platform.openai.com/

##### Setting della chiave
Una volta creata la chiave, abbiamo tre opzioni per importarla, come leggiamo dalla documentazione: https://openai.github.io/openai-agents-python/config/

1 - Utilizzare la funzione  `set_default_openai_key("sk-...")` per settare la chiave in runtime

2 - Salvare la chiave all'interno della variabili d'ambiente del nostro sistema operativo

3 - Sfruttare un file .env per salvare le variabibili a livello di progetto

```python
import os
from dotenv import load_dotenv

load_dotenv()
os.environ["OPENAI_API_KEY"]
```
